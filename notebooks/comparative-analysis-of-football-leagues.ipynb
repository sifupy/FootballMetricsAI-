## Age: 

```python
mean_age_league=df.groupby('Comp')['Age'].mean().reset_index()
plt.figure(figsize=(7,4))
sns.barplot(data=mean_age_league,x='Comp',y='Age')
plt.xlabel("Leagues")
plt.ylabel("Average age")
plt.xticks(rotation=30, ha="right")
plt.title("Average age per league")
plt.show()
```

#### how many player **above|under** average age in each league:

```python
under_25_counts = df[df['Age'] < 25].groupby('Comp').size()
print(under_25_counts)
print("#" * 50)
under_22_counts = df[df['Age'] < 22].groupby('Comp').size()
print(under_22_counts)
```

```python
above_25_counts = df[df['Age'] > 25].groupby('Comp').size()
print(above_25_counts)
print("#" * 50)
above_29_counts = df[df['Age'] > 29].groupby('Comp').size()
print(above_29_counts)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Comp", y="Age", data=df, palette="magma")
plt.xticks(rotation=45)
plt.xlabel("League")
plt.ylabel("AGE")
plt.title("Age distrubution per League")
plt.show()
```

## Nationality:

```python
nationality_counts = df.groupby(["Comp", "Nation"]).size().reset_index(name="count")
nationality_counts
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(data=nationality_counts.sort_values("count", ascending=False).groupby("Comp").head(5),
            x="Nation", y="count", hue="Comp", dodge=False)
plt.xticks(rotation=90)
plt.title("Top Nationalities Frequency per League")
plt.ylabel("Player Count")
plt.xlabel("Nationality")
plt.legend(title="League")
plt.show()
```

##  Performance, Passing, and Defense Summary

```python
# Group by league (Competition)
league_stats = df.groupby("Comp").agg({
    "Gls": "sum",        # Total goals scored
    "Ast": "sum",        # Total assists
    "xG": "mean",        # Average expected goals
    "xAG": "mean",       # Average expected assists
    "Cmp%": "mean",      # Average pass completion
    "PrgP": "mean",      # Progressive passes
    "PrgR": "mean",      # Progressive carries
    "Tkl": "mean",       # Average tackles
    "Int": "mean",       # Average interceptions
    "Clr": "mean",       # Clearances
    "Touches": "mean",   # Average touches per game
}).reset_index()

# Display the table
league_stats
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Gls", y="Comp", data=league_stats, palette="Blues_r")
plt.xlabel("Total Goals")
plt.ylabel("League")
plt.title("Total Goals per League")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Comp", y="Tkl", data=df, palette="magma")
plt.xticks(rotation=45)
plt.xlabel("League")
plt.ylabel("Average Tackles per Player")
plt.title("Defensive Tackles per League")
plt.show()
```

## Players :

```python
# Get youngest player in each league
youngest_players = df.loc[df.groupby("Comp")["Age"].idxmin(), ["Comp", "Player", "Age","Pos"]]

# Get oldest player in each league
oldest_players = df.loc[df.groupby("Comp")["Age"].idxmax(), ["Comp", "Player", "Age","Pos"]]

# Get top goal scorer in each league
top_scorers = df.loc[df.groupby("Comp")["Gls"].idxmax(), ["Comp", "Player", "Gls","Age","Pos"]]

# Get player with highest assist difference (A-xAG) in each league
top_diff_assist = df.loc[df.groupby("Comp")["A-xAG"].idxmax(), ["Comp", "Player", "A-xAG","Age","Pos"]]

# Get player with highest expected goals (xG) in each league
top_xg_players = df.loc[df.groupby("Comp")["xG"].idxmax(), ["Comp", "Player", "xG","Age","Pos"]]
# Get player with highest tackels (Tkl) in each league
top_Tkl_players = df.loc[df.groupby("Comp")["Tkl"].idxmax(), ["Comp", "Player", "Tkl","Age","Pos"]]

# Display the results
print("Youngest Players:\n", youngest_players)
print("\nOldest Players:\n", oldest_players)
print("\nTop Scorers:\n", top_scorers)
print("\nTop Assist Difference Players:\n", top_diff_assist)
print("\nTop xG Players:\n", top_xg_players)
print("\nTop Tackling Players:\n", top_Tkl_players)
```

### Bad finishers :
#### this players should train on thier finishing

```python
# Calculate the difference between expected goals (xG) and actual goals (Gls)
df["Bad_Finishing"] = df["xG"] - df["Gls"]

# Get the players with the highest difference (high xG but low goals)
bad_finishers = df.sort_values("Bad_Finishing", ascending=False)[["Comp", "Player", "xG", "Gls","Pos", "Bad_Finishing"]]

# Display top 10 worst finishers
print(bad_finishers.head(10).to_string(index=False))
```

### Great finishers :
#### players with goals more than thier XG , goats or luck ? 

```python
# Calculate the "luck factor" as Goals - Expected Goals (Gls - xG)
df["Lucky_Finishing"] = df["Gls"] - df["xG"]

# Get the players with the highest positive difference (more goals than expected)
lucky_players = df.sort_values("Lucky_Finishing", ascending=False)[["Comp", "Player", "Gls", "xG","Pos","Lucky_Finishing"]]

# Display top 10 luckiest finishers
print(lucky_players.head(10).to_string(index=False))
```
