### 1ï¸âƒ£ Top Creators Based on Ast_stats_passing, xA, KP, and PPA

```python

top_creators = df[['Player', 'Ast_stats_passing', 'xA', 'KP', 'PPA']]


top_creators['CreativityScore'] = top_creators[['Ast_stats_passing', 'xA', 'KP', 'PPA']].sum(axis=1)
top_creators = top_creators.sort_values(by='CreativityScore', ascending=False).head(10)

top_creators[['Player', 'Ast_stats_passing', 'xA', 'KP', 'PPA', 'CreativityScore']]
```

### 2ï¸âƒ£ Players with Biggest Gap Between xA and Actual Assists

```python
df['xA_gap'] = df['xA'] - df['Ast_stats_passing']

# Top underperformers (xA >> actual assists)
underperformers = df[['Player', 'xA', 'Ast_stats_passing', 'xA_gap']].sort_values(by='xA_gap', ascending=False).head(10)
print("â—ï¸Underperformers (unlucky or teammates missing chances):")
print(underperformers)

# Top overperformers (more assists than expected)
overperformers = df[['Player', 'xA', 'Ast_stats_passing', 'xA_gap']].sort_values(by='xA_gap').head(10)
print("\nâœ¨Overperformers (possible luck or good finishers around):")
print(overperformers)
```

### 3ï¸âƒ£ Relationship Between PrgP, PrgC and Ast_stats_passing

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PrgP', y='Ast_stats_passing', hue='Pos')
plt.title("Progressive Passes vs Assists")
plt.xlabel("Progressive Passes (PrgP)")
plt.ylabel("Assists")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PrgC', y='Ast_stats_passing', hue='Pos')
plt.title("Progressive Carries vs Assists")
plt.xlabel("Progressive Carries (PrgC)")
plt.ylabel("Assists")
plt.grid(True)
plt.show()

# Correlation values
print("Correlation PrgP ~ Assists:", df['PrgP'].corr(df['Ast_stats_passing']))
print("Correlation PrgC ~ Assists:", df['PrgC'].corr(df['Ast_stats_passing']))
```

### Bad finishers :
#### theese players should train on thier finishing

```python
# Calculate the difference between expected goals (xG) and actual goals (Gls)
df["Bad_Finishing"] = df["xG"] - df["Gls"]

# Get the players with the highest difference (high xG but low goals)
bad_finishers = df.sort_values("Bad_Finishing", ascending=False)[["Comp", "Player", "xG", "Gls","Pos", "Bad_Finishing"]]

# Display top 10 worst finishers
print(bad_finishers.head(10).to_string(index=False))
```

### Great finishers :
#### players with goals more than thier XG , goats or luck ? 

```python
# Calculate the "luck factor" as Goals - Expected Goals (Gls - xG)
df["Lucky_Finishing"] = df["Gls"] - df["xG"]

# Get the players with the highest positive difference (more goals than expected)
lucky_players = df.sort_values("Lucky_Finishing", ascending=False)[["Comp", "Player", "Gls", "xG","Pos","Lucky_Finishing"]]

# Display top 10 luckiest finishers
print(lucky_players.head(10).to_string(index=False))
```
