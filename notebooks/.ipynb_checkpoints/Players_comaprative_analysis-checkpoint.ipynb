## üìä Comparative Study Between Players

This analysis explores various aspects of player performance, including transfers, attacking efficiency, creativity, and defensive prowess.

---

### üîÅ **Transferred Players**
- How many players have changed clubs?
- Which nationality has the highest number of transferred players?

---

### ‚öΩ **Attacking Performance**
- **Top Goal Scorers** ‚Äì Who leads in goals scored?
- **Shooting Distance Distribution** ‚Äì How far do players usually shoot from?
- **Efficiency vs Distance** ‚Äì Are players with higher average shot distances less efficient? (`Dist` vs `G/Sh`)
- **Goal Correlation** ‚Äì What other stats are most correlated with goal-scoring?
- **Top Creators** ‚Äì Based on `Ast_stats_passing`, `xA`, `KP`, and `PPA`
- **Under/Overperformers**:
  - Players with a **big gap between xA and actual assists**
  - Players with **high xG but few goals** ‚Üí <span style="color: red">Bad finishers</span>
  - Players with **low xG but many goals** ‚Üí <span style="color: green">Great finishers or lucky</span>

---

### üõ°Ô∏è **Defensive Performance**
1. **Top Tacklers** ‚Äì Who has the most tackles won per 90 minutes?
2. **All-Around Defenders** ‚Äì Best combined `Tkl + Int` players.
3. **Defensive Involvement** ‚Äì Most active players based on `Tkl`, `Int`, `Blocks`, and `Clr`.
4. **Tackles vs Interceptions** ‚Äì What‚Äôs the relationship?
5. **Defensive Errors** ‚Äì Who made the most errors leading to goals?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## ‚öΩ Attacking Performance

```python
df['Pos'].unique()
```

```python
df_attack = df[df['Pos'].str.contains('FW|MF', case=False, na=False)]
df_attack.head()
```

```python
focus_metrics = [
    # Basic Info
    'Player',
    'Pos',
    'Nation',
    'Comp',
    'Age',  
    'MP',              # Matches Played
    'Starts',
    'Min',             # Minutes Played
    '90s',             # Minutes in 90s increments
    
    # Attacking Output
    'Gls', #‚Äì Goals scored
    'Ast', # ‚Äì Assists provided
    'G+A', #‚Äì Goals + Assists
    'xG',  #‚Äì Expected goals
    'xAG', #‚Äì Expected assists
    'npxG', #‚Äì Non-penalty expected goals
    'G-PK', #‚Äì Goals excluding penalties
    'Sh',
    
    #üéØPassing & Creativity Stats
    'PrgP',     #‚Äì Progressive passes
    'PrgC',     #‚Äì Progressive carries
    'KP',         #‚Äì Key passes (passes leading to a shot)
            #‚Äì Pass completion percentage
    'Ast_stats_passing',         #‚Äì Assists
    'xA',            #‚Äì Expected assists
    'PPA',            #‚Äì Passes into the penalty area
    
    # Set Pieces
    'PK',              # Penalty Goals
    'PKatt',           # Penalty Attempts
    
    # Progression
    'PrgR',            # Progressive Receives
    
    #üîÑPossession & Ball Control
    'Touches', #‚Äì Total touches of the ball
    'Carries', #‚Äì Total ball carries
    'PrgR', #‚Äì Progressive runs (carries moving the ball forward significantly)
    'Mis', #‚Äì Miscontrols
    'Dis', #‚Äì Times dispossessed
    
    #üö®Miscellaneous Stats
    'CrdY', #‚Äì Yellow cards
    'CrdR', #‚Äì Red cards
    'PKwon', #‚Äì Penalties won
    'PKcon', #‚Äì Penalties conceded
    'Recov', #‚Äì Ball recoveries
    
    # Advanced Shooting 
    'Sh/90',           # Shots per 90
    'SoT%',            # Shot on Target %
    'G/Sh',            # Goals per Shot
    'Dist',            # Avg Shot Distance
]
```

```python
df_attack=df_attack[focus_metrics]
df_attack.head()
```

### **Correaltion of goals with other stats (which conditions effects the most on scorring)**

```python
corrs=dict(df_attack.corr(numeric_only=True)['Gls'].sort_values(ascending=False))
corrs
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate correlation with 'Gls' and drop NaNs
corrs_series = df_attack.corr(numeric_only=True)['Gls'].dropna()

# Ensure values are scalar floats
corrs_series = corrs_series.apply(lambda x: float(x))

# Sort and get top 10 (excluding 'Gls' itself)
top_10_corrs = corrs_series.drop('Gls', errors='ignore').sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(top_10_corrs.index[::-1], top_10_corrs.values[::-1], color='skyblue')
plt.xlabel('Correlation with Gls')
plt.title('Top 10 Correlated Features with Goals (Gls)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

```python
corrs=dict(df_attack.corr(numeric_only=True)['G-PK'].sort_values(ascending=False))
corrs
```

```python
df = df[df['Sh'] > 0].copy()
df['G_per_Shot'] = df['Gls'] / df['Sh']

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Dist', y='G_per_Shot', hue='Pos')
plt.title("Shot Distance vs Goals per Shot")
plt.xlabel("Avg Shot Distance (Dist)")
plt.ylabel("Goals per Shot (G/Sh)")
plt.grid(True)
plt.show()

# Correlation
print("Correlation between Shot Distance and Goals per Shot:", df['Dist'].corr(df['G_per_Shot']))
```

### 1Ô∏è‚É£ Top Creators Based on Ast_stats_passing, xA, KP, and PPA

```python

top_creators = df[['Player', 'Ast_stats_passing', 'xA', 'KP', 'PPA']]


top_creators['CreativityScore'] = top_creators[['Ast_stats_passing', 'xA', 'KP', 'PPA']].sum(axis=1)
top_creators = top_creators.sort_values(by='CreativityScore', ascending=False).head(10)

top_creators[['Player', 'Ast_stats_passing', 'xA', 'KP', 'PPA', 'CreativityScore']]
```

### 2Ô∏è‚É£ Players with Biggest Gap Between xA and Actual Assists

```python
df['xA_gap'] = df['xA'] - df['Ast_stats_passing']

# Top underperformers (xA >> actual assists)
underperformers = df[['Player', 'xA', 'Ast_stats_passing', 'xA_gap']].sort_values(by='xA_gap', ascending=False).head(10)
print("‚ùóÔ∏èUnderperformers (unlucky or teammates missing chances):")
underperformers


```

```python
# Top overperformers (more assists than expected)
overperformers = df[['Player', 'xA', 'Ast_stats_passing', 'xA_gap']].sort_values(by='xA_gap').head(10)
print("\n‚ú®Overperformers (possible luck or good finishers around):")
overperformers
```

### 3Ô∏è‚É£ Relationship Between PrgP, PrgC and Ast_stats_passing

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PrgP', y='Ast_stats_passing', hue='Pos')
plt.title("Progressive Passes vs Assists")
plt.xlabel("Progressive Passes (PrgP)")
plt.ylabel("Assists")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PrgC', y='Ast_stats_passing', hue='Pos')
plt.title("Progressive Carries vs Assists")
plt.xlabel("Progressive Carries (PrgC)")
plt.ylabel("Assists")
plt.grid(True)
plt.show()

# Correlation values
print("Correlation PrgP ~ Assists:", df['PrgP'].corr(df['Ast_stats_passing']))
print("Correlation PrgC ~ Assists:", df['PrgC'].corr(df['Ast_stats_passing']))
```

### Bad finishers :
#### theese players should train on thier finishing

```python
# Calculate the difference between expected goals (xG) and actual goals (Gls)
df["Bad_Finishing"] = df["xG"] - df["Gls"]

# Get the players with the highest difference (high xG but low goals)
bad_finishers = df.sort_values("Bad_Finishing", ascending=False)[["Comp", "Player", "xG", "Gls","Pos", "Bad_Finishing"]]

# Display top 10 worst finishers
bad_finishers.head(10)
```

### Great finishers :
#### players with goals more than thier XG , goats or luck ? 

```python
# Calculate the "luck factor" as Goals - Expected Goals (Gls - xG)
df["Lucky_Finishing"] = df["Gls"] - df["xG"]

# Get the players with the highest positive difference (more goals than expected)
lucky_players = df.sort_values("Lucky_Finishing", ascending=False)[["Comp", "Player", "Gls", "xG","Pos","Lucky_Finishing"]]

# Display top 10 luckiest finishers
lucky_players.head(10)
```

## üõ°Ô∏èDefensive Performance

```python
focus_metrics = [
    # Basic Info
    'Player',
    'Pos',
    'Nation',
    'Comp',
    'Age',  
    'MP',              # Matches Played
    'Starts',
    'Min',             # Minutes Played
    '90s',             # Minutes in 90s increments
    
    # Attacking Output
    'Gls', #‚Äì Goals scored
    'Ast', # ‚Äì Assists provided
    'KP',         #‚Äì Key passes (passes leading to a shot)
    'PPA',            #‚Äì Passes into the penalty area
    
    #üõ°Ô∏èDefensive Stats
    'Tkl', # ‚Äì Total tackles
    'TklW', # ‚Äì Tackles won
    'Blocks', # ‚Äì Blocks made
    'Int', #‚Äì Interceptions
    'Tkl+Int', # ‚Äì Combined tackles and interceptions
    'Clr', # ‚Äì Clearances
    'Err', # ‚Äì Errors leading to goals
    
  

    
    # Progression
    'PrgR',            # Progressive Receives
    
    #üîÑPossession & Ball Control
    'Touches', #‚Äì Total touches of the ball
    'Carries', #‚Äì Total ball carries
    'PrgR', #‚Äì Progressive runs (carries moving the ball forward significantly)
    'Mis', #‚Äì Miscontrols
    'Dis', #‚Äì Times dispossessed
    
    #üö®Miscellaneous Stats
    'CrdY', #‚Äì Yellow cards
    'CrdR', #‚Äì Red cards
    'PKwon', #‚Äì Penalties won
    'PKcon', #‚Äì Penalties conceded
    'Recov', #‚Äì Ball recoveries
    
]
```

```python
df_defence = df[df['Pos'].str.contains('DF', case=False, na=False)]
df_defence=df_defence[focus_metrics]
df_defence.head()
```

### üõ°Ô∏è 1. Who are the top 10 players with the most tackles won per 90 minutes?

```python
df['TklW_per_90'] = df['TklW'] / df['90s']
top_tacklers = df[['Player', 'Pos', 'TklW', '90s', 'TklW_per_90']].sort_values(by='TklW_per_90', ascending=False).head(10)
print("Top 10 Players by Tackles Won per 90:")
top_tacklers
```

### üîÑ 2. Who are the best all-around defenders (Tkl + Int)?

```python
top_defenders = df[['Player', 'Pos', 'Tkl', 'Int', 'Tkl+Int']].sort_values(by='Tkl+Int', ascending=False).head(10)
print("Top 10 Players by Tackles + Interceptions:")
top_defenders
```

### üö´ 3. Which players are most involved defensively (Tkl, Int, Blocks, Clr) ‚Äì combined total?

```python
df['Def_Activity'] = df['Tkl'] + df['Int'] + df['Blocks'] + df['Clr']
top_def_activity = df[['Player', 'Pos', 'Tkl', 'Int', 'Blocks', 'Clr', 'Def_Activity']].sort_values(by='Def_Activity', ascending=False).head(10)
print("Top 10 Defensively Active Players:")
top_def_activity
```

### üß± 4. What‚Äôs the relationship between Interceptions and Tackles?

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Int', y='Tkl', hue='Pos')
plt.title("Interceptions vs Tackles")
plt.xlabel("Interceptions")
plt.ylabel("Tackles")
plt.grid(True)
plt.show()

# Correlation
print("Correlation between Interceptions and Tackles:", df['Int'].corr(df['Tkl']))
```

### Defensive errors leading to goals

```python
top_errors = df[['Player', 'Pos', 'Err']].sort_values(by='Err', ascending=False).head(10)
print("Players with Most Errors Leading to Goals:")
top_errors
```
