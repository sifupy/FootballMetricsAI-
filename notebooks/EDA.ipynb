```python
focus_metrics = [
    # Basic Info
    'Age',  
    'MP',              # Matches Played
    'Starts',
    'Min',             # Minutes Played
    '90s',             # Minutes in 90s increments
    
    # Attacking Output
    'Gls',             # Goals
    'Ast',             # Assists
    'G+A',             # Goals + Assists
    'G-PK',            # Non-Penalty Goals
    'xG',              # Expected Goals
    'npxG',            # Non-Penalty xG
    'xAG',             # Expected Assisted Goals
    'npxG+xAG',        # Non-Penalty xG + xA
    'G+A-PK',          # Non-Penalty Goals + Assists
    'xG+xAG',          # Total Expected Contribution
    
    # Set Pieces
    'PK',              # Penalty Goals
    'PKatt',           # Penalty Attempts
    
    # Progression
    'PrgC',            # Progressive Carries
    'PrgP',            # Progressive Passes
    'PrgR',            # Progressive Receives
    
    # Discipline
    'CrdY',            # Yellow Cards
    'CrdR',            # Red Cards
    
    # Advanced Shooting (if available)
    'Sh/90',           # Shots per 90
    'SoT%',            # Shot on Target %
    'G/Sh',            # Goals per Shot
    'Dist',            # Avg Shot Distance
]

```

```python
df2=df[focus_metrics]
```

```python
df2.describe()
```

```python
df2.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,12))
sns.heatmap(df2.corr(),cmap='coolwarm',annot=True)
plt.title("Correlations map")
plt.show()
```
